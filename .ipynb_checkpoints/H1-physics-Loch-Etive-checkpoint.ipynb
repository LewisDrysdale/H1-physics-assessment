Here we use the [Pandas](https://pandas.pydata.org/docs/user_guide/index.html) package to read the Excel document into a dataframe called *df*. The data is then displayed as a table below.

**Do not change anything here**

```python
# read file into a dataframe
df = pd.read_excel(filename, sheet_name=0)

#print dataframe to screen
df
```

For ease of working the variables are renamed. The new variable names are: Depth; Temperature; Salinity; and Density. We have to calulate Density using a routine from the [Seawater package](https://pythonhosted.org/seawater/eos80.html).

The variables are extracted from the dataframe using the syntax below

    VARIABLE=df.VARIABLE
    
I have set out the varaibles required from the test data. You may need to change the variable names to match your own data.

**<span style="color: red;">2. Look at the box below and check that the variable names match those in the df table**

```python
# depth
Depth=df.DepSM;

# pressure (required to calculate density)
Pressure=df.PrdM;

# salinity
Salinity=df.Sal00;

# temeperature
Temperature=df.Tv290C;

# this is where density is calculated
Density=sw.eos80.dens(Salinity, Temperature, Pressure)
```

## Excercise 2: Plotting the data

Now we can plot the data using a package called [pyplot](https://matplotlib.org/stable/tutorials/introductory/pyplot.html), which is from matplotlib. You'll see the plot below is 2-D with an X and Y axis. For oceanographic profiles, we usually plot the variable we aere interested in (such as salinity) on the **x** axis and the depth on the **y** axis. 

You'll see hash symbols in the code below. This is a way to write hints and descriptions in the code boxes.

```python
# These are the variables for plotting
x=Salinity;
y=Depth;

# This line calls the figure and creates the axis. ax is now the 'handle' for our axes
fig, ax = plt.subplots();

# This command plots the x and y data on to our axes, ax
ax.plot(x, y,color='k',linestyle='dashed',linewidth=2);
```
